```python
df_city = df.drop (['state/province','country','Date_time','UFO_shape','length_of_encounter_seconds','description','year','month','is_weekend'],axis=1)
df_city
```

```python
df_uniquecity = df_city.drop_duplicates(subset=["city"], keep='first')
df_uniquecity
```

```python
# Create a dictionary containing coordinates
coordinates = {
    "latitude": [29.883056,28.978333],
    "longitude": [
        -97.941111,-96.645833],
    "City": ["san marcos",'edna'],
#     "Population": [2099530, 4637436, 2495825]
}
```

```python
# Create a Pandas DataFrame containing coordinates
coordinates_df = pd.DataFrame(coordinates)

# Display sample data
coordinates_df
```

```python
# Configure the map plot_1
map_plot_1 = coordinates_df.hvplot.points(
    "longitude",
    "latitude",
    geo = True,
    tiles = "OSM",
    scale= 5,
    frame_width = 700,
    frame_height = 500,
    size=5
)

# Display the map plot_1
map_plot_1
```
